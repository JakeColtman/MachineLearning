```python
jet = plt.cm.jet
fig = plt.figure()
x = y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)

plt.subplot(121)
uni_x = stats.uniform.pdf(x, loc=0, scale=5)
uni_y = stats.uniform.pdf(y, loc=0, scale=5)
M = np.dot(uni_x[:, None], uni_y[None, :])
im = plt.imshow(M, interpolation='none', origin='lower',
                cmap=jet, vmax=1, vmin=-.15, extent=(0, 5, 0, 5))

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.title("Landscape formed by Uniform priors.")

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(X, Y, M, cmap=plt.cm.jet, vmax=1, vmin=-.15)
ax.view_init(azim=390)
```

```python
p = pm.Uniform("p", lower = 0, upper = 1)
data = np.array(data)
```

```python
assignment = pm.Categorical("assignment", [p, 1 - p], size=data.shape[0])
```

```python
tau1 = pm.Uniform("tau1", lower = 0, upper = 100)
tau2 = pm.Uniform("tau2", lower = 0, upper = 100)
```

```python
taus = 1.0/pm.Uniform( "stds", 0, 100, size= 2)**2 
```

```python
taus
```

```python
centres = pm.Uniform("means", 0,300, size = 2)
```

```python
centres = pm.Normal("centre", [120, 210], [0.01, 0.01], size = 2)
```

```python
@pm.deterministic
def center_i(assignment=assignment, centres=centres):
    return centres[assignment]
```

```python
@pm.deterministic
def tau_i(assignment=assignment, taus=taus):
    return taus[assignment]
```

```python
observations = pm.Normal("obs", center_i, tau_i, value=data, observed=True)
```

```python
model = pm.Model([p, assignment, observations, taus, centre])
```

```python
mcmc = pm.MCMC( model )
output = mcmc.sample(90000)
```

```python
trace = mcmc.trace("centre").gettrace()
```
