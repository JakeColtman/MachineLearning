Priors can be:

- Objective: Generally flat, allow the posterior to have biggest impact
- Subjective: use previous posteriors or beliefs to shape posterior



```python
%matplotlib inline
import numpy as np
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt
import scipy.stats as stats

figsize(12.5, 3)
colors = ["#348ABD", "#A60628", "#7A68A6", "#467821"]
```

```python
beta = stats.beta.pdf
```

```python
x = np.linspace(0,1) #beta bounded between zero and one
```

```python
y = beta(x, 1, 1)
```

```python
plt.plot(x, y)
```

```python
## Uniform dist is a special case of the beta distribution
```

```python
plt.plot(x, beta(x, 2, 6))
```

```python
def plot_beta(a, b):
    plt.plot(x, beta(x, a, b))
```

```python
plot_beta(10, 1)
```

```python
plot_beta(0.5, 0.5)
```

Bayesian bandit algorithm:

1 - For each bandit, sample from the pror => Xb
2 - Choose the bandit with the highest valued Xb
3 - Observe the result and update the prior on b
4 - GOTO 1

Don't discard losers, but incrementally down weight them
